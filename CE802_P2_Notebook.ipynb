```python
# importing libraries to read the training data and visualise the data
import pandas as pd
import numpy as np
import seaborn as sns

# importing for plot the graphs
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
mydata= pd.read_csv('CE802_P2_Data.csv')
```

```python
#checking the first 10 rows of data
mydata.head(10)
```

```python
# find missing percentage of data in all of the informative predictor
missing_percentage = mydata.isnull().sum()/len(mydata) * 100
print(missing_percentage)
```

```python
# pandas describe() method is used to calculate the statistical values of data fram
mydata.describe()
```

```python
# ploting a histogram for col 21 to check if data is normally distributed or not
mydata.hist(column='F21')
```

```python
# from above graph we can see that data is normally distributed so we
#impute the missing values by mean or median.

#making a copy of original data
mydata_copy = mydata.copy()
```

```python
mydata_copy['F21'] = mydata_copy['F21'].fillna(mydata_copy['F21'].mean())
```

```python
print(mydata_copy)
```

```python
# checking again if there is any missing value
missing_percentage = mydata_copy.isnull().sum()/len(mydata_copy) * 100
print(missing_percentage)
```

```python
# replacing the class values from false and true to 0 and 1

mydata_copy['Class'].replace({False: 0, True: 1}, inplace=True)
mydata_copy
```

```python
#checking correlation among informative predictors
plt.figure(figsize=(24,16))
sns.heatmap(mydata_copy.corr(), cmap="Blues", annot=True)
plt.title('Correlation Matrix')
```

```python
#checking skewness of all the predictor variables
from scipy.stats import skew

for col in mydata_copy:
    print('column name',col)
    print(skew(mydata_copy[col]))
    plt.figure()
    sns.distplot(mydata_copy[col])
    plt.show()
```

```python
# as most of the data is skewed we will get underfitting result
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

x = mydata_copy.drop('Class',axis=1)
y = mydata_copy['Class']
y
```

```python
x_train,x_test, y_train, test_lab = train_test_split(x,y,test_size=0.20, random_state=42)
```

```python
dt_classifier = DecisionTreeClassifier(max_depth =5, random_state = 42)

dt_classifier.fit(x_train, y_train)
```

```python
# visualising decision tree
from sklearn import tree
plt.figure(figsize=(40,20),facecolor = 'lavender')
a = tree.plot_tree(dt_classifier,
                  feature_names = x.columns,
                  class_names = str(y.unique()),
                  rounded = True,
                  filled = True,
                  fontsize = 14)
plt.show()
```

```python
#predict classes on the basis of training data on test data which we split.
test_pred_decision_tree = dt_classifier.predict(x_test)
```

```python
# using confussion matrix to see how the predicted values match up with displaying actual values

y.replace({0:'False',1: 'True'}, inplace=True)
from sklearn import metrics

import seaborn as sns

import matplotlib.pyplot as plt

confusion_matrix = metrics.confusion_matrix(test_lab,  

                                            test_pred_decision_tree)

matrix_df = pd.DataFrame(confusion_matrix)

ax = plt.axes()

sns.set(font_scale=1.3)

plt.figure(figsize=(10,7))

sns.heatmap(matrix_df, annot=True, fmt="g", ax=ax, cmap="magma")

ax.set_ylabel("True Label", fontsize=15)

ax.set_yticklabels(list(y.unique()), rotation = 0)

ax.set_xlabel("Predicted label", fontsize =15)

ax.set_title('Confusion Matrix - Decision Tree')

ax.set_xticklabels([''] + y.unique())

plt.show()
```

```python
# checking accurracy of system for test data
test_pred_decision_tree = dt_classifier.predict(x_test)
print(metrics.classification_report(test_lab, test_pred_decision_tree))
```

```python
# from the above summary I can state that the decision tree for the data given is 86% accurate
# if the max_depth of tree is 5, if we increase or decrease the tree depth the accurracy and precission decreases.
```

```python
from sklearn.ensemble import RandomForestClassifier
rfc = RandomForestClassifier()
rfc.fit(x_train, y_train)

test_predict_rfc = rfc.predict(x_test)
```
